# Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Data/heart_attack_china.csv")
```

```python
print(df.head(5))
```

```python
print(df.describe())
print(df.info())
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
# Firstly we need to check if there are any missing values in the dataset
print(df.isnull().sum())
```

```python
# We have 59617 null values.
print(df['Education_Level'].value_counts())
# Education Level is categorical. Lets set all the null values to a new category called "uknown or none"
df['Education_Level'] = df['Education_Level'].fillna('Unknown or None')
```

```python
# Now lets see if we have fixed the missing values
df.isnull().sum()
```

```python
# Now we can continue the cleaning process by checking for outliers in the dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Function to detect outliers using IQR
def detect_outliers(df):
    outliers = pd.DataFrame(columns=df.columns)
    for col in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = pd.concat([outliers, df[(df[col] < lower_bound) | (df[col] > upper_bound)]])
    return outliers

# Detect outliers
foundOutliers = detect_outliers(df)
print("Number of outliers detected:", foundOutliers.shape[0])
print(foundOutliers)
```

```python
# We didnt find any outliers in the dataset. Now we can check for duplicates in the dataset.
print("Number of duplicates in the dataset:", df.duplicated().sum())
```

```python
cat_featues = ['Gender', 'Smoking_Status', 'Cholesterol_Level',
                        'Air_Pollution_Exposure', 'Physical_Activity', 
                       'Diet_Score', 'Stress_Level', 'Healthcare_Access',
                       'Rural_or_Urban', 'Region', 'Province', 'Hospital_Availability',
                       'Employment_Status', 'Education_Level',
                       'Income_Level', 'Hypertension', 'Diabetes',
                       'Obesity', 'Alcohol_Consumption', 'Family_History_CVD',
                       'TCM_Use', 'Chronic_Kidney_Disease', 
                        'Previous_Heart_Attack', 'Heart_Attack']
df_modeling = pd.get_dummies(df, columns=cat_featues, drop_first=True, dtype=int)
df_modeling.info()
```

```python
# Now we can save the cleaned dataset to a pickle file
import pickle

data = df_modeling
print(data.head(5))
with open('../Data/cleanedDF.pkl', 'wb') as f:
    pickle.dump(data, f)
```
